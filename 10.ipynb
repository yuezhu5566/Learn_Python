交互式环境 (Interactive Environment)
中文解释:
交互式环境是指用户可以实时输入命令并立即看到执行结果的计算环境。在这种环境中,你不需要先写完整个程序再运行,而是可以逐行或逐段输入代码,系统会马上给出反馈。
English:
An interactive environment is a computing environment where users can enter commands in real-time and immediately see the results. Instead of writing a complete program before running it, you can input code line-by-line or section-by-section, and the system provides instant feedback.

```python
def boxPrint(symbol,width,height):
    if len(symbol)!=1:
        raise Exception("Symbol must be a signle character string.")
    if width<=2:
        raise Exception("Width must be greater than 2.")
    if height<=2:
        raise Exception("Height must be greater than 2.")
    print(symbol*width)
    for i in range(height-2):
        print(symbol+(" " *(width-2))+symbol)
        print(symbol*width)

# 注意!这个 for 循环在函数外面,没有缩进!
# Note! This for loop is OUTSIDE the function, no indentation!

for sym,w,h in (("*",4,4),("0", 20, 5),("x",1,3),("ZZ",3,3)):
        try:
            boxPrint(sym,w,h)
        except Exception as err:
            print("An exception happened:"+str(err))
```
