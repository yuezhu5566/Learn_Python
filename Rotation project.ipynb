```python
pythonimport numpy as np

# 创建一个表征矩阵
N = 1000  # Number of samples (细胞数量)
D = 512   # Dimension (神经元数量)

R = np.random.randn(N, D)  # ℝ^(N×D)

print(f"R的形状: {R.shape}")  # (1000, 512)
print(f"R属于: {N}×{D}维实数矩阵")

# 第i个样本的表征 (R_i)
i = 0  # Python索引从0开始，对应数学中的第1个样本
R_i = R[i, :]  # 第i行，所有列
print(f"R_{i}的形状: {R_i.shape}")  # (512,)

# 第j个神经元的激活 (R_{-,j})
j = 0  # 第1个神经元
R_j = R[:, j]  # 所有行，第j列
print(f"R_{{-,{j}}}的形状: {R_j.shape}")  # (1000,)
```
