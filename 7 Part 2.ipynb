```python
import re

def is_strong_password(password: str):
    # def is_strong_password(password: str)-> bool:
    # """检查密码是否为强口令"""
    
    # 长度检查
    if len(password) < 8:
        return False
    # 至少一个大写字母
    if not re.search(r'[A-Z]', password):
        return False
    # 至少一个小写字母
    if not re.search(r'[a-z]', password):
        return False
    # 至少一个数字
    if not re.search(r'[0-9]', password):
        return False
    return True

# 测试
print(is_strong_password("abc"))          # False
print(is_strong_password("Abcdef12"))     # True
print(is_strong_password("password123"))  # False（缺大写）
```

```python
import re

def regex_strip(text: str, char: str = None) -> str:
    """使用正则实现 strip 功能"""
    if char is None:
        # 去除首尾空白
        return re.sub(r'^\s+|\s+$', '', text)
    else:
        # 去除首尾指定字符
        pattern = rf'^[{re.escape(char)}]+|[{re.escape(char)}]+$'
        return re.sub(pattern, '', text)

# 测试
print(regex_strip("   hello   "))          # "hello"
print(regex_strip("xxhelloxx", "x"))       # "hello"
print(regex_strip("!!good!!", "!"))        # "good"
```
