# Writing files

If you perform some calculation, there are a number of reasons why you should store these values somewhere. 

There are three primary ways to store data: raw text, comma separated values (csv), and json.

Typically you would use raw text if you just have text (think like a story) or you want to have some thing like our roster files initially came in.

A CSV file is still plain text, but it is formatted as if it were a spreadsheet. Typically we write a header at the top and every column is separated by a comma (that's where the name comes from). This way when we read it back in, we know to split each row whenever a comma appears. Make sure to not use commas in your values though!

The last way is a JSON [(it's an acronym for something boring)](http://json.org/). JSON files let us store python dictionaries as text! When we read/write files it goes instantly from the raw text to a python dictionary.

```python
import os

# 第一步：读取文件，找到我们需要的行
roster_path = 'Data/Day2-Collections-and-Files/Roster/'
file_name = 'Agatha_Bailey_798.txt'  # 根据你实际的文件名调整
file_path = os.path.join(roster_path, file_name)

# 1a & 1b: 读取文件，找到 name_line 和 height_line
name_line = None
height_line = None

with open(file_path, 'r') as f:
    for line in f:
        if line.startswith('Name:'):
            name_line = line
        elif line.startswith('Height:'):
            height_line = line

print("找到的行:")
print(f"Name line: {name_line}")
print(f"Height line: {height_line}")

# 
```

```python
第二步：分解这些行，提取我们需要的变量
# 2a: 提取 name
name = name_line.split(':', 1)[1].strip()
print(f"\n姓名: {name}")

# 2b & 2c: 提取 height_component_feet 和 height_component_inches
height_string = height_line.split(':', 1)[1].strip()  # 例如: "6ft,0in"
print(f"原始身高字符串: {height_string}")

# 移除 'ft' 和 'in'，然后分割
height_clean = height_string.replace('ft', '').replace('in', '')
height_parts = height_clean.split(',')

height_component_feet = int(height_parts[0].strip())
height_component_inches = int(height_parts[1].strip())

print(f"英尺部分: {height_component_feet}")
print(f"英寸部分: {height_component_inches}")

# 第三步：计算总英寸数
# 3a: total_height_inches
total_height_inches = height_component_feet * 12 + height_component_inches
print(f"\n总英寸数: {total_height_inches} inches")

# 第四步：转换为厘米
# 1 inch = 2.54 cm
total_height_cm = total_height_inches * 2.54
print(f"身高(厘米): {total_height_cm:.2f} cm")

# 第五步：写入新文件
output_file = 'agatha_height.txt'

with open(output_file, 'w') as f:
    # 写入表头
    f.write("Name, Height (cm)\n")
    # 写入数据
    f.write(f"{name}, {total_height_cm:.2f}\n")

print(f"\n✅ 成功写入文件: {output_file}")

 
```
